```python
p=pipeline('automatic-speech-recognition',model='jonatasgrosman/wav2vec2-large-xlsr-53-english')
```

```python
!pip install gradio -q
```

```python
import gradio as gr
import time
```

```python
def transcribe(audio, state=""):
    time.sleep(3)
    text = p(audio)["text"]
    state += text + " "
    return state, state
```

```python
gr.Interface(
    fn=transcribe,
    inputs=[
        gr.inputs.Audio(source="microphone", type="filepath"),
        'state'
    ],
    outputs=[
        "textbox",
        "state"
    ],
    live=True).launch()
```
